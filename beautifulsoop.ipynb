```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://books.toscrape.com/"
```

```python
session = requests.Session()
page = session.get("http://books.toscrape.com/")
```

```python
print(page.status_code)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.title.text)
```

```python
links = soup.find_all('a')                           #this finds all <a> tags

for link in links:
    print(link.get('href'))                          #extracts href
```
